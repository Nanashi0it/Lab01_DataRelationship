# Thư viện

```python
import re
import os
import pandas as pd
import requests
```

# Thu thập dữ liệu

```python
# Lấy nội dung trang web
url = url = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
# Lấy bảng dữ liệu
# NOTE: lấy bảng dữ liệu của 2 ngày trước vì nó đầy đủ hơn
content = re.sub(r'<.*?>', lambda g: g.group(0).upper(), url.text)
data_table = pd.read_html(content)
raw_df = data_table[2]
raw_df
```

```python
# Thông tin của bảng dữ liệu
raw_df.info()
```

```python
# Hiển thị 10 dòng dữ liệu đầu tiên
raw_df.head(10)
```

```python
# Hiển thị 10 dòng dữ liệu cuối cùng
raw_df.tail(10)
```

```python
# Loại bỏ cột và hàng không cần thiết
covid19_statistics_df = raw_df.iloc[8:236, 1:16].reset_index().iloc[:, 1:]
covid19_statistics_df
```

```python
# Loại bỏ dấu '+' và ','
covid19_statistics_df = covid19_statistics_df.replace({'\+':'', ',':''}, regex = True)
covid19_statistics_df
```

```python
# Thông tin của dữ liệu
covid19_statistics_df.info()
```

```python
# Lưu dữ liệu vào file CSV
os.makedirs('./Data', exist_ok= True)
covid19_statistics_df.to_csv('./Data/Covid19_Statistics.csv')
```
